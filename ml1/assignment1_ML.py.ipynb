> AUC (Area Under Curve)-ROC (Receiver Operating Characteristic) is a performance metric, based on varying threshold values, for classification problems. As name suggests, ROC is a probability curve and AUC measure the separability. In simple words, AUC-ROC metric will tell us about the capability of model in distinguishing the classes. Higher the AUC, better the model.<br/>

> ROC – Receiver operating characteristic curve<br/>
Roc curve (False positive rate/1-sensitivity Vs true positive rate/vs 1-sensitivity) allows us to visualize the 
trade-off between the true positive rate and the false positive rate.<br/>
ROC mainly used to evaluate the binary classification accuracy<br/>

>AUC is scale-invariant. It measures how well predictions are ranked, rather than their absolute values.<br/>
AUC is a classification-threshold-invariant. It measures the quality of the model’s predictions irrespective of what classification threshold is chosen.

```python
# Decision threshold & ROC for different thresholds
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import auc

y = [1,0,1,1,0,1]
y_prob = [0.8, 0.96, 0.4, 0.3, 0.2, 0.7]
y1 = [1,1,1,1,1,1] # for threshold value 0
cf_matrix1=confusion_matrix(y,y1)
print("confusion matrix")
print(cf_matrix1)

fpr, tpr, thresholds = roc_curve(y, y1)
print("True positive rate")
print(tpr)
print("False positive rate")
print(fpr)
print("Thresholds")
print(thresholds)
auc_val = roc_auc_score(y, y1)
print('AUC - Test Set: %.2f%%' % (auc_val*100))
print('AUC - '+str(auc(fpr,tpr)))

plt.plot(fpr, tpr, color='orange', label='ROC at y prob 0')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()

y2 = [1,1,1,1,0,1]  # for threshold value 0.2
cf_matrix1=confusion_matrix(y,y2)
print("confusion matrix")
print(cf_matrix1)

fpr, tpr, thresholds = roc_curve(y, y2)
print("True positive rate")
print(tpr)
print("False positive rate")
print(fpr)
print("Thresholds")
print(thresholds)
auc_val = roc_auc_score(y, y2)
print('AUC - Test Set: %.2f%%' % (auc_val*100))
print('AUC - '+str(auc(fpr,tpr)))

plt.plot(fpr, tpr, color='orange', label='ROC at y prob 0.2')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()

y3 = [1,1,1,0,0,1]  # for threshold value 0.3
cf_matrix1=confusion_matrix(y,y3)
print("confusion matrix")
print(cf_matrix1)

fpr, tpr, thresholds = roc_curve(y, y3)
print("True positive rate")
print(tpr)
print("False positive rate")
print(fpr)
print("Thresholds")
print(thresholds)
auc_val = roc_auc_score(y, y3)
print('AUC - Test Set: %.2f%%' % (auc_val*100))
print('AUC - '+str(auc(fpr,tpr)))

plt.plot(fpr, tpr, color='orange', label='ROC at y prob 0.3')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()

y4 = [1,1,0,0,0,1]  # for threshold value 0.4
cf_matrix1=confusion_matrix(y,y4)
print("confusion matrix")
print(cf_matrix1)

fpr, tpr, thresholds = roc_curve(y, y4)
print("True positive rate")
print(tpr)
print("False positive rate")
print(fpr)
print("Thresholds")
print(thresholds)
auc_val = roc_auc_score(y, y4)
print('AUC - Test Set: %.2f%%' % (auc_val*100))
print('AUC - '+str(auc(fpr,tpr)))

plt.plot(fpr, tpr, color='orange', label='ROC at y prob 0.4')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()

y5 = [1,1,0,0,0,0]  # for threshold value 0.4
cf_matrix1=confusion_matrix(y,y5)
print("confusion matrix")
print(cf_matrix1)

fpr, tpr, thresholds = roc_curve(y, y5)
print("True positive rate")
print(tpr)
print("False positive rate")
print(fpr)
print("Thresholds")
print(thresholds)
auc_val = roc_auc_score(y, y5)
print('AUC - Test Set: %.2f%%' % (auc_val*100))
print('AUC - '+str(auc(fpr,tpr)))

plt.plot(fpr, tpr, color='orange', label='ROC at y prob 0.7')
plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

- Different Scenarios with ROC curve and model selection
 - Scneario 1 (Best Scenario)
 - Scenario 2 (Random Guess)
 - Scenario 3 (Worst case)
 - Scenario 4 (Normal scenario)
 
 **Scneario 1 (Best Scenario)**
 For any classification model, the best scenario is when there is a clear distinction between the two / all the classes.
 The graph above shows the Predicted Class Probability for both classes 0 and 1. The threshold is 0.5 which means, if the predicted probability of the class for an instance is less than 0.5, that instance is predicted to be an instance of class 0. If the probability of the class for an instance is equal or greater than 0.5, the instance is classified as the instance of class 1.
 
 <br/><br/>
 
 **Scneario 2 (Random Guess)**
 In the event where both the class distribution simply mimic each other, AUC is 0.5. In other words, our model is 50% accurate for instances and their classification. The model has no discrimination capabilities at all in this case.

 <br/><br/>
 
 <br/><br/>
 **Scneario 3 (Worst Case Scenario)**
 If the model completely misclassifies the classes, it is the worst case.
Completely opposite of the best case scenario (scenario #1), in this case, all the instances of class 1 are misclassified as class 0 and all the instances of class 0 are misclassified as class 1.

 <br/><br/>
 
 <br/><br/>
 **Scneario 4 (Normal Scenario)**
 In a usual industry scenario, best cases are never observed. We never get a clear distinction between the two classes.
 In this case, as observed, we have some overlapping and that introduces Type 1 and Type 2 errors to the model prediction. In this case we get AUC to be somewhere between 0.5 and 1.

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
df=pd.read_csv("dataset.csv")
df.rename({'Predicted Class': 'y_predicted','Actual Class': 'y_actual'},axis=1, inplace = True)
print(df)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report

cf_matrix=confusion_matrix(df['y_actual'],df['y_predicted'], labels=[0,1,2,3])
#confusion_matrix=pd.crosstab(df['y_actual'],df['y_predicted'],rownames=['Actual'],colnames=['Predicted'])
print("\x1b[1;31m"+"confusion matrix"+"\x1b[0m")
print(" Predicted")
print(cf_matrix)

print("\x1b[1;31m"+"manually computed tp,fp,fn,tn"+"\x1b[0m")
print("  TP FP FN TN")
print("0 33 6  16 105")
print("1 32 13 4  111")
print("2 29 10 11 110")
print("3 27 10 8  115")

print("\x1b[1;31m"+"classification report"+"\x1b[0m")
print(classification_report(df['y_actual'],df['y_predicted'],labels=[0,1,2,3],digits=3))
print("\x1b[1;31m"+"overall accuracy"+"\x1b[0m")
print(accuracy_score(df['y_actual'],df['y_predicted']))
print("\x1b[1;31m"+"overall accuracy manual observation"+"\x1b[0m")
print(cf_matrix.diagonal().sum()/cf_matrix.sum()) # for multiclass classification do we need to follow (tp+fp)/(tp+fp+tn+fn) which is giving the different accuracy value?
print("\x1b[1;31m"+"class-wise accuracy using (tp+tn)/(tp+tn+fp+fn)"+"\x1b[0m")
print("0- "+str((33+105)/(33+6+16+105)))
print("1- "+str((32+111)/(33+13+4+111)))
print("2- "+str((29+110)/(33+10+11+110)))
print("3- "+str((27+115)/(33+10+8+115)))

print("\x1b[1;31m"+"precision (tp/(tp+fp))"+"\x1b[0m")
precision0=cf_matrix.diagonal()[0]/cf_matrix[:,0].sum()
precision1=cf_matrix.diagonal()[1]/cf_matrix[:,1].sum()
precision2=cf_matrix.diagonal()[2]/cf_matrix[:,2].sum()
precision3=cf_matrix.diagonal()[3]/cf_matrix[:,3].sum()
print("class "+str(0)+": "+str(precision0))
print("class "+str(1)+": "+str(precision1))
print("class "+str(2)+": "+str(precision2))
print("class "+str(3)+": "+str(precision3))

print("\x1b[1;31m"+"recall (tp/(tp+fn))"+"\x1b[0m")
recall0=cf_matrix.diagonal()[0]/cf_matrix[0].sum()
recall1=cf_matrix.diagonal()[1]/cf_matrix[1].sum()
recall2=cf_matrix.diagonal()[2]/cf_matrix[2].sum()
recall3=cf_matrix.diagonal()[3]/cf_matrix[3].sum()
print("class "+str(0)+": "+str(recall0))
print("class "+str(1)+": "+str(recall1))
print("class "+str(2)+": "+str(recall2))
print("class "+str(3)+": "+str(recall3))
#print(cf_matrix.diagonal()/cf_matrix.sum(axis=1))

print("\x1b[1;31m"+"macro average"+"\x1b[0m")
print(precision_score(df['y_actual'],df['y_predicted'], average='macro'))
print("\x1b[1;31m"+"macro average manual observation"+"\x1b[0m")
total=0.0
for i in range(0,4):
    total=total+cf_matrix.diagonal()[i]/cf_matrix[:,i].sum()
print(total/4)

print("\x1b[1;31m"+"weighted average"+"\x1b[0m")
print(precision_score(df['y_actual'],df['y_predicted'], average='weighted'))
print("\x1b[1;31m"+"weighted average manual observation"+"\x1b[0m")

total1=0.0
total2=0.0
for i in range(0,4):
    #(weight1*precision1+weight2*precision2)/(weight1+weight2)
    total1=total1+(cf_matrix[:,i].sum()*(cf_matrix.diagonal()[i]/cf_matrix[:,i].sum()))
    total2=total2+cf_matrix[:,i].sum()
print(total1/total2)

print("\x1b[1;31m"+"f1 score"+"\x1b[0m")
print("class 0: "+str(2*((recall0*precision0)/(recall0+precision0))))
print("class 1: "+str(2*((recall1*precision1)/(recall1+precision1))))
print("class 2: "+str(2*((recall2*precision2)/(recall2+precision2))))
print("class 3: "+str(2*((recall3*precision3)/(recall3+precision3))))

print("\x1b[1;31m"+"type1 error fp/(tn+fp)"+"\x1b[0m")
print("class 0: "+str(6/(6+105)))
print("class 1: "+str(13/(13+111)))
print("class 2: "+str(10/(10+110)))
print("class 3: "+str(10/(10+115)))

print("\x1b[1;31m"+"type2 error fn/(fn+tp)"+"\x1b[0m")
print("class 0: "+str(16/(16+33)))
print("class 1: "+str(4/(4+32)))
print("class 2: "+str(11/(11+29)))
print("class 3: "+str(8/(8+27)))
```
